## **2.1. Теоретический материал – Условный оператор**
---
**If – Условный оператор**

В коде часто приходится проверять выполнимость или
невыполнимость каких-то условий. Синтаксис:
```
if <условие1 (булевское выражение)> :
    <код, который выполнится, если условие верно>
elif <условие2 (булевское выражение)>:
    <код, который выполнится, если условие1 было неверно, а условие2 верно>
else:
    <код, который выполнится, если условие1 и условие2 были неверны>
```
Обратите внимание, что код, который должен выполняться внутри каждого условия, записывается с отступом в 4 пробела от уровня if, elif и else: в питоне области видимости переменных обозначаются отступами.

**То есть, отступы позволяют понять, где начинается код, который должен выполняться при выполнении условия в if, и где заканчивается.**

## **3.1. Теоретический материал – Циклы, массивы, срезы**
---
**Циклы - for и while**

Цикл **while** проверяет истинность условия, и если условие истинно, то выполняет инструкции цикла.

Синтаксис:
```
while условие:
    инструкции
```
Инструкции, которые относятся к циклу while, располагаются на последующих строках и должны иметь отступ от начала ключевого слова while.

Цикл **for** проходит по набору значений, помещает каждое значение в переменную с которой в цикле можно производить различные действия.

Синтаксис:
```
for переменная in набор_значений:
    инструкции
```
После ключевого слова **for** идет название переменной, в которую
будут помещаться значения. После оператора **in** указывается набор значений
и двоеточие. Инструкции, которые относятся к циклу, располагаются на
последующих строках и должны иметь отступ.

При создании цикла удобно пользоваться функцией **range(a, b)**,
которая создает последовательность чисел от **a** до **b-1**.

### **Задача:**
Для функции $\left| \sin{(xe^{\cos(x)})} \right|$ на интервале [0;10] с шагом 0,05 построить график функции, закрасить площадь под ним и вычислить значение этой площади.

Для вычисления площади воспользуйтесь функцией trapz(), вычисляющей интеграл по правилу трапеции. Для ее корректной работы необходимо подключить следующие библиотеки:
```
from scipy.integrate import simps
from numpy import trapz
```
### **Решение:**

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz
x = np.arange(0.0, 10, 0.1)
y = np.abs(np.sin(x*np.exp(np.cos(x))))
plt.grid()
plt.plot(x, y, c = "r")
plt.fill_between(x, y)

area = trapz(y)
print(area)
````

### **Задача:**
Задайте массив случайных значений из интервала (0; 1).

Рассчитайте средние и медианные значения для массива, сравните результаты, какие выводы можно сделать о значениях?

Постройте точечную диаграмму рассеяния полученного ряда.
### **Решение:**

````python
import random as ran
import matplotlib.pyplot as pyp
count_elem = int(input('Введите число элементов в массиве: '))
array_r = []
# В задании не указано каким способом расчитывать - вручную
# или через функции библиотеки
sred = 0
for i in range (0, count_elem):
    array_r.append(ran.random())
    sred += array_r[i]
sred /= count_elem
for i in range (0, count_elem-1):
    for j in range (0, count_elem-1-i):
        if array_r[j] > array_r[j+1]:
            help = array_r[j]
            array_r[j] = array_r[j+1]
            array_r[j+1] = help
med = count_elem // 2
if count_elem / 2 == med:
    med = (array_r[med] + array_r[med-1])/2
else:
    med = array_r[med]
print('Среднее значение:', sred, '| Медианное значение:', med)
# Не совсем понял о каком рассеивании говорится ведь
# отклонение в средних значениях, а не на каждом элементе
x = []
for i in range (0, count_elem):
    x.append(i+1)
y = []
if count_elem% 2 == 1:
    for i in range(0, count_elem // 2):
        y.append(0)
    y.append(sred)
else:
    for i in range(0, count_elem // 2 - 1):
        y.append(0)
    y.append(sred)
    y.append(sred)
for i in range(count_elem // 2 + 1, count_elem):
    y.append(0)
pyp.grid()
pyp.scatter(x, array_r)
pyp.scatter(x, y)
# О выводах: значения отличаются как правило не более чем
# на 10% объём выборки не влияет на отличие (при малом обзоре)
# поскольку числа генерируются случайным образом. Формально
# при увеличении числа элементов, значения должны сближаться
# Среднее значение может сильно отличаться от мидианного, если
# число элементов не на много больше числа "ошибок"
````

````python
import math as m
def calcul(fun, x, y):
    if fun == 1:
        return x+y
    elif fun == 2:
        return x-y
    elif fun == 3:
        return x*y
    elif fun == 4:
        while y == 0:
            print('Ошибка, на ноль нельзя делить')
            y = float(input('Введите y: '))
        return x/y
    elif fun == 5:
        return m.exp(x+y)
    elif fun == 6:
        return m.sin(x+y)
    elif fun == 7:
        return m.sin(x+y)
    elif fun == 8:
        while (x == 0) & (y < 0):
            print('Ошибка, основание степени не может равняться нулю при отрицательном показателе')
            x = float(input ('Введите x: '))
        while (x < 0) & (int(y) != y):
            print('Ошибка, основание степени не может быть отрицательным при дробном показателе')
            x = float(input ('Введите x: '))
        return m.pow(x, y)
    else:
        return 'Ошибка'
menu = -1
x = 'check'
y = 'check'
while menu != 0:
    print('0. Закончить работу')
    print('1. x+y')
    print('2. x-y')
    print('3. x*y')
    print('4. x/y')
    print('5. e^(x+y)')
    print('6. sin(x+y)')
    print('7. cos(x+y)')
    print('8. x^y')
    menu = int(input('Введите номер меню для выполнения операции: '))
    if (menu <= 0) | (menu > 8):
        continue
    if x == 'check':
        x = float(input('Введите число x: '))
    if y == 'check':
        y = float(input('Введите число y: '))
    result = calcul(menu, x, y)
    print(result)
    help1 = int(input('Сохранить результат, как x? Да(1), нет(0) '))
    if help1:
        x = result
    else:
        x = 'check'
    help2 = int(input('Сохранить результат, как y? Да(1), нет(0) '))
    if help2:
        y = result
    else:
        y = 'check'
  # Простой калькулятор с элементарными проверками
  # Проверять ввод строки вместо числа не стал
````
